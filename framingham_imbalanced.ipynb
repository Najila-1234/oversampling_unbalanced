```python
import pandas as pd
df=pd.read_csv('/content/framingham.csv')
df
```

```python
df.isna().sum()
```

```python
df['cigsPerDay']=df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
df
```

```python
df['BPMeds']=df['BPMeds'].fillna(df['BPMeds'].mode()[0])
df
```

```python
df['totChol']=df['totChol'].fillna(df['totChol'].mean())
df
```

```python
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
df['heartRate']=df['heartRate'].fillna(df['heartRate'].mean())
df['glucose']=df['glucose'].fillna(df['glucose'].mean())
df
```

```python
df.isna().sum()
```

```python
df.drop(['education'],axis=1, inplace=True)
df
```

```python
df['TenYearCHD'].value_counts()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
yo.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_new=sc.transform(x_train)
x_test_new=sc.transform(x_test)
```

```python
# use knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_new,y_train)
y_pred=knn.predict(x_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train_new,y_train)
y_pred_nb=nb.predict(x_test_new)
y_pred_nb
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_nb))
```

```python
# undersampling
from imblearn.under_sampling import RandomUnderSampler
under=RandomUnderSampler()
xu,yu=under.fit_resample(x,y)
```

```python
yu.value_counts()
```

```python
xu.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xu,yu,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train_new=scalar.transform(x_train)
x_test_new=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_new,y_train)
y_pred=knn.predict(x_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
